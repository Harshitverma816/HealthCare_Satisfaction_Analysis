```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
df.columns
```

```python
df.drop(columns=[
    'Patient_Survey_Star_Rating_Footnote',
    'HCAHPS_Answer_Percent_Footnote',
    'Number_of_Completed_Surveys_Footnote',
    'Survey_Response_Rate_Percent_Footnote'
], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['ZIP_Code'] = df['ZIP_Code'].astype(str)
```

```python
numeric_cols = ['Patient_Survey_Star_Rating', 
                'HCAHPS_Answer_Percent', 
                'HCAHPS_Linear_Mean_Value', 
                'Number_of_Completed_Surveys', 
                'Survey_Response_Rate_Percent']

for col in numeric_cols:
    df[col] = (df[col].astype(str)
               .str.replace('%', '', regex=False)
               .str.replace(',', '', regex=False)
               .replace(['Not Available', 'Not Applicable', 'N/A', 'nan', 'NaN', 'None', ' ', 'Blanks'], None)
               .astype(float))
```

```python
df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce')
df['End_Date'] = pd.to_datetime(df['End_Date'], errors='coerce')
```

```python
df['Number_of_Completed_Surveys'] = df['Number_of_Completed_Surveys'].fillna(0).astype(int)
```

```python
# Fill missing values within each State group by mean ofthat State
df['Patient_Survey_Star_Rating'] = df.groupby('State')['Patient_Survey_Star_Rating'].transform(
    lambda x: x.fillna(round(x.mean(), 2))
).round(2)

df['HCAHPS_Linear_Mean_Value'] = df.groupby('State')['HCAHPS_Linear_Mean_Value'].transform(
    lambda x: x.fillna(round(x.mean(), 2))
).round(2)

df['HCAHPS_Answer_Percent'] = df.groupby('State')['HCAHPS_Answer_Percent'].transform(
    lambda x: x.fillna(round(x.mean(), 2))
).round(2)

df['Number_of_Completed_Surveys'] = df.groupby('State')['Number_of_Completed_Surveys'].transform(
    lambda x: x.fillna(round(x.mean(), 2))
).round(2)

df['Survey_Response_Rate_Percent'] = df.groupby('State')['Survey_Response_Rate_Percent'].transform(
    lambda x: x.fillna(round(x.mean(), 2))
).round(2)
```

```python
print(df.dtypes)
```

```python
# Surveys Duration
df['Survey_Period_Days'] = (df['End_Date'] - df['Start_Date']).dt.days
```

```python
# Response Rate Category
def categorize_response(rate):
    if pd.isna(rate):
        return 'Unknown'
    elif rate < 30:
        return 'Low'
    elif 30 <= rate < 60:
        return 'Medium'
    else:
        return 'High'

df['Survey_Response_Category'] = df['Survey_Response_Rate_Percent'].apply(categorize_response)
```

```python
# Satisfaction Level (categorical)
def satisfaction_label(rating):
    if pd.isna(rating):
        return 'Unknown'
    elif rating < 2.5:
        return 'Poor'
    elif 2.5 <= rating < 3.5:
        return 'Average'
    elif 3.5 <= rating < 4.5:
        return 'Good'
    else:
        return 'Excellent'

df['Satisfaction_Level'] = df['Patient_Survey_Star_Rating'].apply(satisfaction_label)
```

```python
# Rating to Response Ratio
df['Rating_to_Response_Ratio'] = df['Patient_Survey_Star_Rating']/ df['Survey_Response_Rate_Percent']
```

```python
# Performance Score
df['Performance_Score'] = (0.6 * df['Patient_Survey_Star_Rating']) + (0.4 * (df['Survey_Response_Rate_Percent'] / 20))
```

```python
# Question Category 
def map_question_category(question):
    if 'nurse' in question.lower():
        return 'Nurse Communication'
    elif 'doctor' in question.lower():
        return 'Doctor Communication'
    elif 'clean' in question.lower():
        return 'Cleanliness'
    elif 'quiet' in question.lower():
        return 'Environment'
    elif 'pain' in question.lower():
        return 'Pain Management'
    elif 'discharge' in question.lower():
        return 'Discharge Info'
    elif 'medic' in question.lower():
        return 'Medication Explanation'
    else:
        return 'Other'

df['Question_Category'] = df['HCAHPS_Question'].apply(map_question_category)
```

```python
# Extract Survey Year
df['Survey_Year'] = df['Start_Date'].dt.year
```

```python
# Top Performer Flag
df['Top_Performance'] = np.where(df['Patient_Survey_Star_Rating'] >= 4.0, 1, 0)
```

```python
# Preview new columns
print(df[['Facility_Name', 'State', 'Patient_Survey_Star_Rating', 
          'Survey_Response_Rate_Percent', 'Satisfaction_Level', 
          'Survey_Response_Category', 'Performance_Score', 
          'Question_Category', 'Survey_Year']].head())
```

```python
from tqdm import tqdm
from sqlalchemy import create_engine
import time

start_time = time.time()

print("Uploading data in chunks...")

engine = create_engine("mysql+pymysql://root:@localhost/healthcare_analysis")

chunk_size = 1000
for i in tqdm(range(0, len(df), chunk_size)):
    df.iloc[i:i+chunk_size].to_sql(
        'healthcare_survey',
        con=engine,
        if_exists='append' if i > 0 else 'replace',
        index=False
)

end_time = time.time()
elapsed_time = end_time - start_time

print("Upload complete successfully")
print(f"Total Time taken: {elapsed_time: .2f} seconds ({elapsed_time/60:.2f} minutes)")
```

## Queries

```python
-- Q1 - Count the total number of Hospitals (unique facilities)
select count(distinct Facility_ID) as Total_Hospital 
from healthcare_survey;

-- Q2 - Find Top 10 states with the most hospitals
select State, 
count(distinct Facility_ID) as total_facilities 
from healthcare_survey 
group by State 
order by total_facilities desc 
limit 10;

-- Q3 - Average patient survey star rating by state 
select State, 
round(avg(Patient_Survey_Star_Rating), 2) as avg_star_rating 
from healthcare_survey 
group by State 
order by avg_star_rating desc;

-- Q4 - Calculate Average performance score by satisfaction level 
select Satisfaction_Level, round(avg(Performance_Score), 2) as avg_performance   
from healthcare_survey 
group by Satisfaction_Level 
order by avg_performance desc;

-- Q5 - Identify hospitals with rating below average of their state 
select h1.Facility_Name, h1.State, h1.Patient_Survey_Star_Rating
from healthcare_survey h1
join (
    select State, avg(Patient_Survey_Star_Rating) as state_avg
    from healthcare_survey
    group by State
) h2 on h1.State = h2.State
where h1.Patient_Survey_Star_Rating < h2.state_avg
order by h1.State, h1.Patient_Survey_Star_Rating asc;

-- Q6 - Compare satisfaction by question category 
select Question_Category,
       round(avg(Patient_Survey_Star_Rating), 2) as avg_rating,
       round(avg(HCAHPS_Answer_Percent), 2) as avg_answer_percent
from healthcare_survey
group by Question_Category
order by avg_rating desc;

-- Q7 - Average surveys completed & response rate per state 
select State,
round(avg(Number_of_Completed_Surveys), 0) as avg_completed_surveys,
round(avg(Survey_Response_Rate_Percent), 2) as avg_response_rate
from healthcare_survey
group by State
order by avg_response_rate desc;

-- Q8 - Find top 10 facilities with the highest star ratings 
select Facility_Name,
round(avg(Patient_Survey_Star_Rating), 2) as Avg_Star_Rating
from Healthcare_survey
group by Facility_Name
order by Avg_Star_Rating desc
limit 10;

-- Q9 - Analyze Satisfaction by State
select State,
round(avg(HCAHPS_Answer_Percent), 2) as Avg_Satisfaction,
count(distinct Facility_ID) as Num_Facilities
from Healthcare_survey
group by State
order by Avg_Satisfaction desc;

-- Q10 - Year Wise Performance Trend
select Survey_Year,
round(avg(HCAHPS_Linear_Mean_Value), 2) as Avg_Linear_Score,
round(avg(Patient_Survey_Star_Rating), 2) as Avg_Star_Rating
from healthcare_survey
group by Survey_Year
order by Survey_Year;

-- Q11 - Satisfaction Level Distribution
select Satisfaction_Level,
count(*) as Num_Responses,
round(avg(HCAHPS_Answer_Percent), 2) as Avg_Satisfaction
from Healthcare_survey
group by Satisfaction_Level desc;

-- Q12 - Facilities with best rating-to-response ratio
select Facility_Name,
round(avg(Rating_to_Response_Ratio), 2) as Avg_Ratio
from Healthcare_survey
group by Facility_Name
order by Avg_Ratio desc 
limit 10;

-- Q13 - Identify underperforming facilities (below threshold)
select Facility_Name, State,
avg(Patient_Survey_Star_Rating) as Avg_Rating
from Healthcare_survey
group by Facility_Name, State
having Avg_Rating < 2.5
order by Avg_Rating asc;
```
